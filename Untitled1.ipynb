```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

# Download stopwords (hanya sekali)
nltk.download('stopwords')

# Inisialisasi stopwords dan stemmer
stop_words = set(stopwords.words('english'))
stemmer = SnowballStemmer("english")

# Fungsi preprocessing
def preprocess_text(text):
    # Lowercase
    text = text.lower()

    # Hapus angka dan tanda baca
    text = re.sub(r'[^a-z\s]', '', text)

    # Tokenisasi
    tokens = text.split()

    # Hapus stopwords
    tokens = [word for word in tokens if word not in stop_words]

    # Stemming
    tokens = [stemmer.stem(word) for word in tokens]

    # Gabungkan lagi jadi string
    return ' '.join(tokens)

# Terapkan fungsi ke kolom Text â†’ jadi kolom clean_text
df['clean_text'] = df['Text'].apply(preprocess_text)

# Lihat contoh hasil
df[['Text', 'clean_text']].head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
import nltk
nltk.download('stopwords')

# === 1. Sampling data untuk efisiensi (gunakan jika dataset besar) ===
sampled_df = df.sample(5000, random_state=42)

# Ambil kolom clean_text hasil preprocessing
texts = sampled_df['clean_text'].values

# === 2. TF-IDF Vectorization ===
vectorizer = TfidfVectorizer(max_df=0.8, min_df=5, stop_words='english')
X = vectorizer.fit_transform(texts)

print(f"Bentuk matriks TF-IDF: {X.shape}")

# === 3. LSA dengan Truncated SVD ===
n_topics = 10  # jumlah topik yang ingin ditemukan
lsa_model = TruncatedSVD(n_components=n_topics, random_state=42)
lsa_topic_matrix = lsa_model.fit_transform(X)

print(f"Bentuk hasil LSA: {lsa_topic_matrix.shape}")

# === 4. Tampilkan 10 kata teratas dari tiap topik ===
terms = vectorizer.get_feature_names_out()

print("\nTopik dan kata-kata terpenting:\n")
for i, comp in enumerate(lsa_model.components_):
    top_terms = [terms[idx] for idx in comp.argsort()[-10:]]
    print(f"Topik {i+1}: {' | '.join(top_terms)}")

# === 5. Visualisasi Distribusi Topik Dominan ===
topic_labels = lsa_topic_matrix.argmax(axis=1)

plt.figure(figsize=(8, 5))
plt.hist(topic_labels, bins=n_topics, edgecolor='black')
plt.title("Distribusi Topik Dominan")
plt.xlabel("Topik")
plt.ylabel("Jumlah Dokumen")
plt.xticks(range(n_topics))
plt.show()
```

```python
# ========================
# 1. Import library utama
# ========================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

nltk.download('stopwords')

# ========================
# 2. Baca data
# ========================
df = pd.read_csv('/content/drive/MyDrive/Datasets/Reviews.csv')  # ganti path jika berbeda

# ========================
# 3. Bersihkan data
# ========================
df = df.dropna()
df = df[df['Score'] != 3]  # buang review netral
df['Positivity'] = np.where(df['Score'] > 3, 1, 0)  # buat label binary

# ========================
# 4. Preprocessing teks
# ========================
stop_words = set(stopwords.words('english'))
stemmer = SnowballStemmer('english')

def preprocess_text(text):
    text = text.lower()
    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)
    text = re.sub(r"\d+", "", text)
    tokens = text.split()
    tokens = [stemmer.stem(word) for word in tokens if word not in stop_words]
    return " ".join(tokens)

df['clean_text'] = df['Text'].apply(preprocess_text)

# ========================
# 5. Siapkan fitur dan label
# ========================
X = df['clean_text']
y = df['Positivity']

# ========================
# 6. Split data latih dan uji
# ========================
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# ========================
# 7. TF-IDF Vectorization
# ========================
tfidf = TfidfVectorizer(max_df=0.8, min_df=5, stop_words='english')
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

# ========================
# 8. Latih model klasifikasi
# ========================
model = LogisticRegression()
model.fit(X_train_tfidf, y_train)

# ========================
# 9. Evaluasi model
# ========================
y_pred = model.predict(X_test_tfidf)

print("Classification Report:\n", classification_report(y_test, y_pred))

# Visualisasi Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Negatif', 'Positif'], yticklabels=['Negatif', 'Positif'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
